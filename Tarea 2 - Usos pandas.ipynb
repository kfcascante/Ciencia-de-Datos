```python
#Cargar datos del dataset desde un archivo de Excel

data  = pd.read_csv ('titanic.csv')
data.head()


# como saber cuantos resgitrso tiene un data set

print (data.shape)

```

```python
# como contar los datos y ver cuales estan nulos

# tomar e cuenta que el resultado de de personas en la tabla es de 418 y en algunos caso no todos tienen valores
print (data.count())

# otra forma de ver los valores nulos se hace de la siguiente forma
print("-------------------------------")

col_names = data.columns.tolist()
for column in col_names:
    print ("Valores nulos en <{0}>:{1}".format(column, data[column].isnull().sum()))
    
    



```

```python
    
# Como  estandarizar datos, en este caso vamos a pasar el sexo a letras  Male y Female a M y F

# creamos un diccionario con los valores originales 

d = {'male' : 'M', 'female' : 'F'}

#Utilizamos un lambda para el remplazo, en una sola linea n.n

data['Sex'] = data['Sex'].apply(lambda x:d[x])

#Revisar el cambio

data['Sex'].head()


#Forma de acceder a las columnas


data.Age
```

```python
# agrupar por 2 campos

pd.crosstab(data.Age, data.Sex)
```

```python
#Agrupar por varias columnas


pclass_gender_age_sex = data.groupby (['Pclass', 'Sex'])['Age'].sum()
pclass_gender_age_sex
```

```python
#Habra alguna relacion entre tipo de ticket y edad aqui vamosa graficar como lo podemos hacer.

import matplotlib.pyplot as plt

fig = plt.figure(figsize=(20, 10))

for t_class in [1,2,3] :
    data.Age[data.Pclass == t_class].plot(kind='kde')
    plt.legend(("1ra. Clase", "2da. Clase", "3ra.Clase"))
plt.show()
```
